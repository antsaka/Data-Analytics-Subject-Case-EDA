```python
col_names=['Biaya_Perawatan','Kapasitas','Ukuran_Bagasi','Keamanan','Status_Terjual']
for col in col_names:
  print(df[col].value_counts())
```

```python
from sklearn import preprocessing
le_Kapasitas=preprocessing.LabelEncoder()
le_Kapasitas.fit(['Dua','Empat','Lebih'])
df['Kapasitas']=le_Kapasitas.transform(df['Kapasitas'])
```

```python
from sklearn import preprocessing
le_Biaya_Perawatan=preprocessing.LabelEncoder()
le_Biaya_Perawatan.fit(['Rendah','Sedang','Tinggi'])
df['Biaya_Perawatan']=le_Biaya_Perawatan.transform(df['Biaya_Perawatan'])
```

```python
from sklearn import preprocessing
le_Ukuran_Bagasi=preprocessing.LabelEncoder()
le_Ukuran_Bagasi.fit(['Kecil','Sedang','Besar'])
df['Ukuran_Bagasi']=le_Ukuran_Bagasi.transform(df['Ukuran_Bagasi'])
```

```python
from sklearn import preprocessing
le_Keamanan=preprocessing.LabelEncoder()
le_Keamanan.fit(['Rendah','Sedang','Tinggi'])
df['Keamanan']=le_Keamanan.transform(df['Keamanan'])
```

```python
X=df.drop(['Status_Terjual'],axis=1)
y=df['Status_Terjual']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train , y_test = train_test_split(X , y, test_size=0.15 , random_state = 42)
X_train.shape, X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
Tree=DecisionTreeClassifier(criterion='entropy')
Tree.fit(X_train,y_train)
```

```python
predTree_train=Tree.predict(X_train)
hasil_train=pd.DataFrame ({'Predicted_Status_Terjual':predTree_train,'Actual_Status_Terjual':y_train})
print(hasil_train)
```

```python
predTree_test=Tree.predict(X_test)
hasil_test=pd.DataFrame ({'Predicted_Status_Terjual':predTree_test,'Status_Terjual':y_test})
print(hasil_test)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
cnfMatrix_train=confusion_matrix(y_train,predTree_train)
print ('Decision Tree Accuracy:',metrics.accuracy_score(y_train,predTree_train))
cnfMatrix_train
```

```python
cnfMatrix_test=confusion_matrix(y_test,predTree_test)
print ('Decision Tree Accuracy:',metrics.accuracy_score(y_test,predTree_test))
cnfMatrix_test
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train,predTree_train))
```

```python
print(classification_report(y_test,predTree_test))
```

```python
from sklearn import tree
import graphviz
dot_data = tree.export_graphviz(Tree,out_file=None,feature_names=X_train.columns,class_names=np.unique(y_train),filled=True,rounded=True,special_characters=True)
graph=graphviz.Source(dot_data)
graph
```
